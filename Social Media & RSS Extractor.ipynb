```python
def getPosts(soup,json):
    listOfPosts = soup.find('div',{'class':'entity-all feed-container-theme ember-view'})
    LinkedInArticle = ""
    #print(listOfPosts)
    allPosts = listOfPosts.findAll('span',{'class':'break-words'})
    for i in allPosts:
        spanText = i.find('span',{'dir':'ltr'})
        LinkedInArticle = LinkedInArticle + spanText.text
        #print(spanText.text)
    print("LinkedIn Article for this company", LinkedInArticle)
    json.update({"LinkedIn Article":LinkedInArticle})
    print("#%#%@#%@^#%@#^%&*#@%&#%@&*%@*&#%@*&#%@*&#%&*#^&#*^&*#^@&#^&*^&*#^&*#^@*&^#@&*#^*#@&^#&*")
```
